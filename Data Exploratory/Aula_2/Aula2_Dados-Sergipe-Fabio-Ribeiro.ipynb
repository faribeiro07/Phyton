## 5.5. Exercício - Medidas de posição

Faça o download dos dados de [nascidos vivos em 2017 de Sergipe](https://drive.google.com/file/d/1-JnSCtkibxdKWxkrWSml2HD60PejMm0s/view?usp=sharing) e responda às seguintes perguntas:



1.   Qual a média da idade da mãe, peso, apgar 1 min, apgar 5 min, quantidade de filhos vivos e quantidade de filhos mortos?

```
sinasc_se.IDADEMAE = pd.to_numeric(sinasc_se.IDADEMAE).astype(pd.Int64Dtype())
sinasc_se.PESO = pd.to_numeric(sinasc_se.PESO).astype(pd.Int64Dtype())
sinasc_se.APGAR1 = pd.to_numeric(sinasc_se.APGAR1).astype(pd.Int64Dtype())
sinasc_se.APGAR5 = pd.to_numeric(sinasc_se.APGAR5).astype(pd.Int64Dtype())
sinasc_se.QTDFILVIVO = pd.to_numeric(sinasc_se.QTDFILVIVO).astype(pd.Int64Dtype())
sinasc_se.QTDFILMORT = pd.to_numeric(sinasc_se.QTDFILMORT).astype(pd.Int64Dtype())

round(sinasc_se[['IDADEMAE','PESO', 'APGAR1', 'APGAR5', 'QTDFILVIVO', 'QTDFILMORT']].mean(),2)
```

2.   Qual a idade tal que metade das idades das mães são menores que ela (mediana)?

```
round(sinasc_se.IDADEMAE.median(),2)
```

3.   Qual o apgar de 1 min que mais aparece (moda)?

```
sinasc_se.APGAR1.mode()
```

4.   Qual o apgar de 5 minutos que mais aparece?

```
sinasc_se.APGAR5.mode()
```

5.   Qual o peso da criança tal que 10% dos pesos são menores (P10)?

```
sinasc_se['PESO'].quantile([0.1])
```

6.   Qual o peso da criança tal que 75% dos pesos são menores (Q3)?

```
percentis = sinasc_se['PESO'].quantile([0.75])
lista_percentis = percentis.values
lista_percentis
```

7.   Qual a quantidade de filhos mortos que mais aparece?

```
sinasc_se.QTDFILMORT.mode()
```

8.   Qual a média de idade de mães solteiras?

```
round(sinasc_se[sinasc_se['ESTCIVMAE'] == '1']['IDADEMAE'].mean(),0)
```

9.   Qual a quantidade de filhos vivos das mães com até 3 anos de escolaridade tal que 25% dessa quantidade são menores (Q1)?

```
sinasc_se[(sinasc_se['ESCMAE'] <= '3')]['QTDFILVIVO'].quantile([0.25])
```
