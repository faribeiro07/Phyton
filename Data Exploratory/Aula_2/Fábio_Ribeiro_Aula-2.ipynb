```
import pandas as pd
```

```
lista_altura =[1.70, 1.65,1.83]
lista_sexo = ['f','f','m']
lista_peso = [70, 65, 83]

df = pd.DataFrame(lista_altura, columns=['Altura'])
df = pd.DataFrame({'Altura': lista_altura, 'Sexo': lista_sexo, 'Peso': lista_peso})
df
```

```
#resumo dos dados
df.describe()
```

```
#como adicionar todos os registros qualitativos e quantitativos
df.describe(include = 'all')
```

```
#Calculo da média da altura
edia = df['Altura'].mean()
media
```
